# Task 3: Feature Engineering

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv("Dataset .csv")
df
```

### 1. Extract additional features from the existing columns, such as the length of the restaurant name or address.

```python
df["Restaurants Name's Length"] = df["Restaurant Name"].apply(lambda x : len(x))
df["Address Length"] = df["Address"].apply(lambda x : len(x))

df.head()
```

### 2. Create new features like "Has Table Booking" or "Has Online Delivery" by encoding categorical variables.

```python
Encoding_Dict = {"Yes":1, "No": 0}

df["Has Table Booking encoding"] = df["Has Table booking"].map(Encoding_Dict)
df["Has Online Delivery encoding"] = df["Has Online delivery"].map(Encoding_Dict)

df.head()
```

### Conclusion

```python
# 1. Created 2 new columns for the length of each restaurants and address name.
# 2. Created two more binary columns that contain encoding of 'has online delivery' and 'table booking' where 0 means 'NO' 
# and 1 menas 'Yes'  
```
