### 2. Explore the distribution of categorical variables like "Country Code," "City," and "Cuisines." Identify the top cuisines and cities with the highest number of restaurants. 

```python
# Using seaborn library, ploted the count plot for country code
# we are instructing sns.countplot() to order the bars according to the index labels of the Top Country code
sns.countplot(data = df, x = df['Country Code'],palette="dark")
plt.title("Top 10 Distribution of Restaurants by Country code")
plt.ylabel("No of Restaurants")
plt.show()
```

```python
# Using seaborn library, ploted the count plot to count restaurants in Top 10 Cities 
# we are instructing sns.countplot() to order the bars according to the index labels of the Top_cities
Top_cities = df['City'].value_counts().head(10)
sns.countplot(data = df, order = Top_cities.index,y = df['City'],palette="dark")
plt.title("Top 10 Cities with Highest Number of Restaurants")
plt.show()
```

```python
# Using seaborn library, ploted the count plot for Top 10 Cuisines 
# we are instructing sns.countplot() to order the bars according to the index labels of the Top_cuisines
Top_Cuisines = df['Cuisines'].value_counts().head(10)
sns.countplot(data = df, order = Top_Cuisines.index, y =df['Cuisines'],palette="dark")
plt.title("Top 10 Cuisines with Highest Number of Restaurants")
plt.xlabel("Number of Restaurants")
plt.ylabel("Names of the Cuisines")
plt.show()
```
