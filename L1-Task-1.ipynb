### 2. Check for missing values in each column and handle them accordingly.

```python
# TO check which column has how many missing values !!
missing_values = df.isnull().sum()
missing_values
# Through this we got to know that the cuisines column has 9 missing values
```

```python
# To know the intial data of cuisines 
s = df["Cuisines"].head()
s
```

```python
# There are several v=ways to deal with missing values: Dropping rows, fill Unknown or Unavailable, I choose Imputation with 
# mode of the column which has missing values 
Value_Mode = df["Cuisines"].mode()[0]
Value_Mode
df["Cuisines"] = df["Cuisines"].fillna(Value_Mode)
df.isnull().sum()
# Now we can see no null values present in it, null values from the cuisines column has been handled 
```

### 3. Perform data type conversion if necessary. Analyze the distribution of the target variable("Aggregate rating") and identify any class imbalances.

```python
df.dtypes
```

```python
# Counting each rating accordingly of the column Aggregate rating
Count_Rating = df["Aggregate rating"].value_counts().sort_index()
Count_Rating
```

```python
# Finding the class imbalance by counting rating divide by total count rating 
Class_Imbalance = Count_Rating / Count_Rating.sum()
Class_Imbalance
```
