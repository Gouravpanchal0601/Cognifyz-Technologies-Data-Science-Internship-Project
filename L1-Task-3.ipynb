# Task 3 : Geospatial Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import geopandas as gpd
import matplotlib.pyplot as plt
```

### 1. Visualize the locations of restaurants on a map using latitude and longitude information.

```python
# Load the dataset
df = pd.read_csv('Dataset .csv')

# Created point geometries from latitude and longitude using gpd.points_from_xy
geometry = gpd.points_from_xy(df.Longitude, df.Latitude)

# Geodf containing the restaurant data along with the geometry
Geodf = gpd.GeoDataFrame(df, geometry=geometry)

# Now here load the world map using naturalearth_lowres
Earth = gpd.read_file(gpd.datasets.get_path('naturalearth_lowres'))

# Here the red sports are the restaurants location on earth
fig, ax = plt.subplots(figsize=(14, 12))

# Here ploted the world map using Earth.plot
Earth.plot(ax=ax)

# Here plot the Restaurant locations using Geodf.plot
Geodf.plot(ax=ax, color='red', markersize=10, label='Restaurant Locations')
plt.legend()
plt.show()
```

### 2. Analyze the distribution of restaurants across different cities or countries. 

```python
Top_Cities = df['City'].value_counts().head(10).index
# Top_Cities
sns.countplot(data=df, y = df['City'], order = Top_Cities,palette='dark')
plt.title("Top 10 Cities with Highest Number of Restaurants")
plt.xlabel("Number of Restaurants")
plt.plot()
```

### 3. Determine if there is any correlation between the restaurant's location and its rating.

```python
Correlation = df[['Latitude', 'Longitude', 'Aggregate rating']].corr()
Correlation
sns.heatmap(Correlation,annot=True,fmt=".2f")
plt.show()
```
