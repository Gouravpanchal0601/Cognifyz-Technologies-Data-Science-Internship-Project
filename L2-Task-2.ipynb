# Task 2: Price Range Analysis

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv("Dataset .csv")
df
```

### 1. Determine the most common price range among all the restaurants.

```python
# All common price ranges among all restaurants 
Common_Price = df['Price range'].mode()[0]

# printing the same
print(f"Most Common Price range among all the Restaurants is : {Common_Price}")
```

### 2. Calculate the average rating for each price range.

```python
# Calculating the average rating for each price range
Avg_Rating = df.groupby('Price range')['Aggregate rating'].mean()

# Printing the same
print(f"Average rating for each price range is : {Avg_Rating}")
```

### 3. Identify the color that represents the highest average rating among different price ranges.

```python
# For geting the index from the Avg_rating which has the highest value that is 4th index which has the highest rating 
High_avg_rating = Avg_Rating.idxmax()

# Avg_Rating.plot(kind="bar")
# plt.axhline(Avg_Rating[High_avg_rating],color='red',linestyle="--",label = "Highest Avg Rating")

# Creating the bar plot
plt.bar(Avg_Rating.index, Avg_Rating)

# Puting the bar which has the highest rating into red 
plt.bar(High_avg_rating,Avg_Rating[High_avg_rating],color='red')

# Setting the name for x and y labels
plt.title("Highest Average Rating by among different price ranges")
plt.xlabel("Price Ranges")
plt.xticks(rotation =0)
plt.ylabel("Average Rating")
plt.show()
```
