# Task 1: Table Booking and Online Delivery

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv("Dataset .csv")
df
```

### 1. Determine the percentage of restaurants that offer table booking and online delivery.

```python
# Finding length of the total restaurants of dataset
Total_Restaurants = len(df)
Total_Restaurants

# Taking out the percentage of restaurants for online delivery and for table booking
Online_Delivery = (len(df[df['Has Online delivery']=='Yes']))/Total_Restaurants*100
Table_Booking = (len(df[df['Has Table booking']=='Yes']))/Total_Restaurants*100

# Printing the same
print(f"Percentage of restaurants offer Online delivery :{Online_Delivery:.2f}%")
print(f"Percentage of restaurants offer Table booking :{Table_Booking:.2f}%")
```

### 2. Compare the average ratings of restaurants with table booking and those without. 

```python
# Taking out the average rating of restaurants which has table booking and not have table booking 
Avg_TableBooking = df[df["Has Table booking"]=='Yes']['Aggregate rating'].mean()
Avg_Not_TableBooking = df[df["Has Table booking"]=='No']['Aggregate rating'].mean()

# Printing the same
print(f"Average rating of Restaurants with Booking : {Avg_TableBooking:.2f}%")
print(f"Average rating of Restaurants without booking : {Avg_Not_TableBooking:.2f}%")
```

### 3. Analyze the availability of online delivery among restaurants with different price ranges.

```python
# Taking out the online delivery value among different price range
Delivery_By_Price = df.groupby('Has Online delivery')['Price range'].value_counts().unstack()

# Ploting the bar
Delivery_By_Price.plot(kind='bar',figsize=(8,6))

# Giving the name of x and y labels 
plt.title('Availability of online delivery among restaurants with different price range')
plt.ylabel('Proportion of Restaurants')
plt.xticks(rotation=0)
plt.plot()
```
