# Task 3: Data Visualization

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns

df = pd.read_csv("Dataset .csv")
df
```

### 1. Create visualizations to represent the distribution of ratings using different charts (histogram, bar plot, etc.)

```python
# 1 Distribution of rating using Histogram 
plt.figure(figsize=(8, 6))
plt.hist(df['Aggregate rating'],bins=30,edgecolor='black')
plt.title("Distributon of Aggregate rating using Histogram")
plt.xlabel("Aggregate ratings")
plt.ylabel("Number of Restaurants")
plt.show()
```

```python
# 2 Distribution of rating using Bar Graph 
Rating_counts = df['Aggregate rating'].value_counts().sort_index()

# PLoting the graph
plt.figure(figsize=(8, 6))
plt.bar(Rating_counts.index,Rating_counts.values,edgecolor="black",width =0.06)

# Setting the name of the label
plt.title("Distributon of Aggregate rating using Bar Graph")
plt.xlabel("Aggregate Ratings")
plt.ylabel("Number of Restaurants")
plt.show()
```

### 2. Compare the average ratings of different cuisines or cities using appropriate visualizations.

```python
# Comparing the average rating of Top 20 cusines with highest ratings that is 4.9
Avg_rating_cuisines = df.groupby("Cuisines")['Aggregate rating'].mean().nlargest(20)
Avg_rating_cuisines

# plt.barh(avg_rating.index ,avg_rating.values)
sns.barplot(x=Avg_rating_cuisines.values, y=Avg_rating_cuisines.index, orient='horizontal')
plt.title("Top 20 Cuisines with Highest Average Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Top 20 Cuisines")
plt.show()
```

```python
# Comparing the average rating of Top 20 Cities with highest ratings that is 4.9
Avg_rating_city = df.groupby("City")['Aggregate rating'].mean().nlargest(20)
Avg_rating_city

# plt.barh(avg_rating.index ,avg_rating.values)
sns.barplot(x=Avg_rating_city.values, y=Avg_rating_city.index, orient='horizontal')
plt.title("Top 20 Cities with Highest Average Ratings")
plt.xlabel("Average Rating")
plt.ylabel("Top 20 Cities")
plt.show()
```

### 3. Visualize the relationship between various features and the target variable to gain insights.

```python
# Scatter plot to visualize the relationship between 'Average Cost for two' and 'Aggregate rating'
sns.scatterplot(x='Average Cost for two', y='Aggregate rating', data=df)
plt.title('Relationship between Average Cost for two and Aggregate rating')
```

```python
# Scatter plot to visualize the relationship between 'Votes' and 'Aggregate rating'
sns.scatterplot(x='Votes', y='Aggregate rating', data=df)
plt.title('Relationship between Votes and Aggregate rating')
```

```python
sns.boxplot(x='Price range', y='Aggregate rating', data=df)
```

```python
sns.boxplot(x='Has Table booking', y='Aggregate rating', data=df)
```

```python
sns.boxplot(x='Has Online delivery', y='Aggregate rating', data=df)
```

```python
# Visualize the relationship between various features and the target variable by using pair plot
sns.pairplot(data=df, vars = ['Aggregate rating','Votes','Average Cost for two'])
plt.suptitle("Relationship between various features and target variable")
plt.show()
```
